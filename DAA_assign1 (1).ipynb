```python
import random
import time
import matplotlib.pyplot as plt

# Linear Search function
def linear_search(arr, key):
    for i in range(len(arr)):
        if arr[i] == key:
            return i
    return -1

# Size of the array
n = 10000
arr = [random.randint(1, 10**6) for _ in range(n)]

# Insert special values for known positions
best_case_key = 999999
arr[0] = best_case_key

average_case_key = 888888
arr[n // 2] = average_case_key

worst_case_key = -1  # Not in array

# Time best case
start = time.perf_counter()
linear_search(arr, best_case_key)
end = time.perf_counter()
best_time = (end - start) * 1000  # Convert to milliseconds

# Time average case
start = time.perf_counter()
linear_search(arr, average_case_key)
end = time.perf_counter()
average_time = (end - start) * 1000  # ms

# Time worst case
start = time.perf_counter()
linear_search(arr, worst_case_key)
end = time.perf_counter()
worst_time = (end - start) * 1000  # ms

# Print actual times
print("Linear Search Time Analysis (n = {})".format(n))
print(f"Best Case Time: {best_time:.4f} ms")
print(f"Average Case Time: {average_time:.4f} ms")
print(f"Worst Case Time: {worst_time:.4f} ms")

# Plotting the graph
cases = ['Best Case', 'Average Case', 'Worst Case']
times = [best_time, average_time, worst_time]

plt.figure(figsize=(8, 5))
plt.bar(cases, times, color=['green', 'orange', 'red'])
plt.xlabel('Case Type')
plt.ylabel('Time (milliseconds)')
plt.title('Linear Search Performance Analysis')
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import random
import matplotlib.pyplot as plt

# Linear search function that counts comparisons
def linear_search(arr, key):
    comparisons = 0
    for i in range(len(arr)):
        comparisons += 1
        if arr[i] == key:
            return i, comparisons
    return -1, comparisons

# Input sizes to test
input_sizes = [100, 500, 1000, 5000, 10000]

# To store comparisons for each case
best_case_comparisons = []
average_case_comparisons = []
worst_case_comparisons = []

# Run for each input size
for n in input_sizes:
    arr = [random.randint(1, 100000) for _ in range(n)]

    # Set known keys
    arr[0] = 111111  # Best case
    arr[n // 2] = 222222  # Average case
    # Worst case: key not present
    key_not_found = -1

    # Best Case
    _, best = linear_search(arr, 111111)

    # Average Case
    _, avg = linear_search(arr, 222222)

    # Worst Case
    _, worst = linear_search(arr, key_not_found)

    # Store comparisons
    best_case_comparisons.append(best)
    average_case_comparisons.append(avg)
    worst_case_comparisons.append(worst)

# Tabulate results
print(f"{'Input Size':<12}{'Best Case':<12}{'Avg Case':<12}{'Worst Case'}")
for i in range(len(input_sizes)):
    print(f"{input_sizes[i]:<12}{best_case_comparisons[i]:<12}{average_case_comparisons[i]:<12}{worst_case_comparisons[i]}")

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(input_sizes, best_case_comparisons, 'go--', label='Best Case (O(1))')
plt.plot(input_sizes, average_case_comparisons, 'bo--', label='Average Case (O(n/2))')
plt.plot(input_sizes, worst_case_comparisons, 'ro--', label='Worst Case (O(n))')

plt.title("Linear Search: Number of Comparisons vs Input Size")
plt.xlabel("Input Size (n)")
plt.ylabel("Number of Comparisons")
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()

     
```
